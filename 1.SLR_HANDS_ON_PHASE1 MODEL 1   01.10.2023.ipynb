```python
# pandas is a library created in python programming contains functions
import pandas as pd
```

```python
#dataset is a key function to collect data from .csv file (comma seperated value file)
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#running vaiable to obtain input and output data
dataset
```

```python
# split input data from dataset
independent=dataset[["YearsExperience"]]
```

```python
#run the assigned variable independent
independent
```

```python
#split output data from dataset
dependant=dataset[["Salary"]]
```

```python
#run the assigned the vaiable
dependant
```

```python
#use sklearn library for split the train set and test set from whole dataset 70:30
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split (independent,dependant,test_size=0.30, random_state=0)
```

```python
# running the assigned variable X_train (Input Training Data)
X_train
```

```python
# running the assigned variable Y_train (Output Training Data)
Y_train
```

```python
# running the assigned variable X_test (Input Test Data)
X_test
```

```python
# running the assigned variable Y_test (Output Test Data)
Y_test
```

```python
#creating the model using sklearn library for given ip and op
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
#to find the weight w or slope m as read in y=wx+b in SLR,Machine learning
weight=regressor.coef_
weight
```

```python
#to find the bais b or Origin b as read in y=wx+b in SLR,Machine learning
bais=regressor.intercept_
bais
```

```python
#to calculate predicated output data
Y_pred=regressor.predict(X_test)
```

```python
# to evaluation of model / metrics
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_pred)
```

```python
#run the variable to check value, if it is 1, best, if it is 0, poor.
r_score
```

```python
#to create file name
import pickle
filename=("finalized_model_linear.sav")
```

```python
#to put the files into the new file name
pickle.dump(regressor,open(filename,'wb'))
```

```python
#deployment stage to run the model with variable inputs
loaded_model=pickle.load(open("finalized_model_linear.sav", 'rb'))
result=loaded_model.predict([[10]])
```
